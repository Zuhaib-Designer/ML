**Task1:**

Write a program that takes salary as input. Using conditional statements, calculate the final tax
based on the following rules:                                                                                        (Marks 3)


• If the salary is less than 30,000 → Tax rate is 5%


• If the salary is between 30,000 and 70,000 → Tax rate is 15%  


• If the salary is greater than 70,000 → Tax rate is 25%

```python
# Taking salary as an input

salary = int(input("Enter your salary: " ))

if salary<30000:
  tax = salary * (0.05)
  print(f"Tax on {salary} salary is {tax}")
elif salary >= 30000 and salary <= 70000:
  tax = salary * (0.15)
  print(f"Tax on {salary} salary is {tax}")
else:
  tax = salary * (0.25)
  print(f"Tax on {salary} salary is {tax}")
```

**Task2:**

Given a list of words:                                                   (Marks 3)

words = ["apple", "banana", "kiwi", "cherry", "mango"]           

Create a dictionary that maps each word to its corresponding length.


Example Output: ({"apple": 5, "banana": 6, "kiwi": 4, "cherry": 6, "mango": 5})

```python
# Given list of words
words = ["apple", "banana", "kiwi", "cherry", "mango"]

dic = {}

for i in range(len(words)):
  dic[words[i]] = len(words[i])

print(dic)
```

**Task3:**

Create a Chat System using Object-Oriented Programming (OOP) concepts.             (Marks 4)

You need to create the following classes:

• User  
• Message  
• ChatRoom  

The system should implement the following functionalities:

• Sending messages  
• Viewing chat history  
• User joining and leaving the chat room

```python
class User:
    def __init__(self, name):
      self.name = name

class Message:
    def __init__(self, sender, text):
        self.sender = sender
        self.text = text
    def show(self):
      return self.sender.name + ":" + self.text

class ChatRoom:
    def __init__(self):
        self.users = []
        self.messages = []


    def join(self, user):
      self.users.append(user)
      print(user.name, "Joined the chat room")

    def leave(self, user):
      self.users.remove(user)
      print(user.name, "Leave the chat room")


    def sent_message(self, user, text):
      mes = Message(user, text)
      self.messages.append(mes)

    def display_history(self):
      print("\n********Chat History*********")
      for m in self.messages:
        print(m.show())


```

```python
chat = ChatRoom()

User1 = User("Babar")
User2 = User("Faheem")

chat.join(User1)
chat.join(User2)

chat.sent_message(User1, "Hello! yar kasa ho")
chat.sent_message(User2, "ma allah ka sukar thek tu suna aj kal konsa match khail raha ha?")

chat.display_history()

chat.leave(User1)
```
