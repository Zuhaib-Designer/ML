# Task 1

Make a program in which user enter his total marks and obtained marks then display student is pass or
fail.

```python
# Take the input from the user as a total_marks and Obtained Marks
total_marks = int(input("Enter Your total Marks :"))
Obtained_marks = int(input("Enter your obtained Marks :"))

# To calculte the criteria
calculated_criteria = (Obtained_marks/total_marks) * 100

# Standarized the criteria
standard_criteria = 50.00

# To make decision about student will pass or fail against Standard Criteria
if calculated_criteria < standard_criteria :
  print(f"Student fail with {calculated_criteria:.2f}% against the standard criteria 50.00%")
else:
  print(f"Student pass with {calculated_criteria:.2f}% against the standard criteria 50.00%")
```

# Task 2
Make a calculator which suggest your grades on the behalf of marks. (Hint 80 :A, 70:B, 60:C, 50:D)

```python
# Take the input from user as a marks
marks = int(input("Enter marks to calculate grades :"))

# To make the decision based on the given Standard
if marks>= 80:
  print("Student pass with 'A' grades")
elif marks<80 and marks >= 70:
  print("Student pass with 'B' grades")
elif marks<70 and marks >= 60:
  print("Student pass with 'C' grades")
elif marks<60 and marks >= 50:
  print("Student pass with 'D' grades")
```
