**Task9:**

Enter
your matriculation marks subject wise and the system will display
total marks and percentage of all also display subject wise
percentage


````python
# Enter the total number of the subject

total_subject = int(input("Enter the total Number of the subject "))
subject = []
total_marks = 0
for i in range(total_subject):
  subject_marks = float(input(f"Enter the marks of subject{i+1} "))
  subject.append(subject_marks)

for j in range(total_subject):
  if subject[j]>150:
    print("OOPs! Maximum marks should be 150")
    subject.pop(j)
  else:
    percentage = (subject[j]/150) * 100
    print(f"Percentage of subject{j+1} is {percentage:0.4f}%")

for k in range(total_subject):
  total_marks = total_marks + subject[k]

print(f"Total Marks are {total_marks:0.4f}")




````
